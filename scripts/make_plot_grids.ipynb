```python
%load_ext autoreload
%autoreload 2

import numpy as np
from os import listdir
import matplotlib.pyplot as plt
from pendulibrary.integrate import integrate_state
from pendulibrary.plotters import plot_timeline_grid
```

```python
fams = [f.removesuffix(".npz") for f in listdir("../database")]
```

```python
for fname in fams:
    data = np.load(f"../database/{fname}.npz")
    x0s = data["x0s"]
    periods = data["periods"]
    tangents = data["tangents"]
    Lr, Mr = data["params"]
    ts, xs, fs = integrate_state(x0s[-1], periods[-1], Lr, Mr, 1e-14)
    fig = plot_timeline_grid(xs, ts, fs, Lr)
    fig.savefig(f"../plots/{fname}_end.png", bbox_inches="tight")
    plt.close(fig)
```
