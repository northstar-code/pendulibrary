```python
from pathlib import Path
from multiprocessing.pool import ThreadPool
import subprocess
```

```python
mp4_root = Path("../plots/mp4s")
gif_root = Path("../plots/gifs")
```

```python
def convert(mp4):
    some_name = mp4.parent.name
    gif = gif_root / f"{some_name}-{mp4.stem}.gif"
    if gif.exists():
        return 0
    gif.parent.mkdir(parents=True, exist_ok=True)

    subprocess.run(
        [
            "ffmpeg",
            "-i",
            str(mp4),
            # "-vf",
            # "fps=30,split[s0][s1];[s0]palettegen=max_colors=4[p];[s1][p]paletteuse",
            "-vf",
            "split[s0][s1];[s0]palettegen[p];[s1][p]paletteuse",
            "-loop",
            "0",
            str(gif),
        ],
        check=True,
    )
```

```python
with ThreadPool() as pool:
    pool.map(convert, list(mp4_root.rglob("*.mp4")))
```
