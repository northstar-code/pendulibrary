```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm.auto import tqdm

from pendulibrary.integrate import integrate_state

int_tol = 1e-14
```

```python
Lr = 1.0
Mr = 1.0
Tf = 500.
```

```python
trajs = []
for p in tqdm(np.linspace(3.0, 0.5, 64, False)[::-1]):
    #         t1(0) = t2(0) = 0, dt1(0) = -dt1(0) = p
    x0 = np.array([0.0, 0.0, -p, p])
    ts, xs, _ = integrate_state(x0, Tf, Lr, Mr, 1e-14)
    y = -(np.cos(xs[0]) + Lr * np.cos(xs[1]))
    x = np.sin(xs[0]) + Lr * np.sin(xs[1])
    trajs.append((x.copy(), y.copy()))
```

```python

fig, axs = plt.subplots(8, 8, figsize=(15, 15))
axf = axs.ravel()

for i in range(64):
    x, y = trajs[i]
    axf[i].plot(x, y, lw=0.1)
    axf[i].axis("equal")
    axf[i].set(xticks=[], yticks=[])
fig.tight_layout()
fig.supxlabel("")
plt.show()
```
