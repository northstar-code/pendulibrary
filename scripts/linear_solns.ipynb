```python
import numpy as np
import matplotlib.pyplot as plt

from pendulibrary.common import eom, get_A_raw
from pendulibrary.integrate import integrate_state, integrate_state_stm
from pendulibrary.interpolate import interp_hermite

from scipy.linalg import expm
from numpy import pi
```

```python
Lr = 1.0
Mr = 1.0
```

```python
Xref = np.array([0.0, 0.0, 0.0, 0.0])
A = get_A_raw(Xref, Lr, Mr)  # linearized dynamics near origin

# both frequencies
w1, w2 = np.sqrt(2 + np.sqrt(2)), np.sqrt(2 - np.sqrt(2))
# periods
T1, T2 = 2 * np.pi / w1, 2 * np.pi / w2

# Solutions for first frequency
Phi1 = expm(A * T1)
eigs1 = np.linalg.eig(Phi1)
X01 = (eigs1.eigenvectors[:, -1] + eigs1.eigenvectors[:, -2]).real


# Solutions for first frequency
Phi2 = expm(A * T2)
eigs2 = np.linalg.eig(Phi2)
X02 = (eigs2.eigenvectors[:, -1] + eigs2.eigenvectors[:, -2]).real

step = 1e-3
```

```python
# interpolate to plot smoother (spline, faster)
ts_out, xs_out, fs_out = integrate_state(X01 * step, T1, 1e-12, Lr, Mr)
t1, x1 = interp_hermite(ts_out, xs_out, fs_out, np.linspace(0.0, T1, 100))


# interpolate to plot smoother (spline, faster)
ts_out, xs_out, fs_out = integrate_state(X02 * step, T2, 1e-12, Lr, Mr)
t2, x2 = interp_hermite(ts_out, xs_out, fs_out, np.linspace(0.0, T2, 100))
```
